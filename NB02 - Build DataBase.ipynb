```python
import kaggle
import pandas as pd
import csv
import os
import json
from lets_plot import *
from lets_plot.mapping import *
LetsPlot.setup_html()
#from plotnine import *
from dotenv import dotenv_values
```

```python
download_path = r'data_proj'
csv_file = [file for file in os.listdir(download_path) if file.endswith('.csv')][0]
csv_path = os.path.join(download_path, csv_file)
df1 = pd.read_csv(csv_path)
df1.shape
```

```python
df1.head()
```

```python
df1=df1.drop(columns=['CrimeCode','Inside/Outside','Weapon','Post','District','Location 1','Premise'])
df1['City'] = 'Blatimore'
df1
```

```python
df2 = pd.read_csv('data_proj/crime.csv')
df2.head()
```

```python
df2['CrimeDate'] = pd.to_datetime(df2['DAY'].astype(str) + '/' + df2['MONTH'].astype(str) + '/' + df2['YEAR'].astype(str), format='%d/%m/%Y')
df2['CrimeDate'] = df2['CrimeDate'].dt.strftime('%d/%m/%Y')
df2 = df2.dropna(subset=['HOUR', 'MINUTE'])
df2['HOUR'] = df2['HOUR'].astype(int).astype(str).str.zfill(2)
df2['MINUTE'] = df2['MINUTE'].astype(int).astype(str).str.zfill(2)
df2['CrimeTime'] = pd.to_datetime(df2['HOUR'] + ':' + df2['MINUTE'], format='%H:%M').dt.time
df2=df2.drop(columns=['YEAR','MONTH','DAY','HOUR','MINUTE','X','Y'])
df2['Total Incidents'] = 1
df2['City'] = 'Vancouver'

df2
```

```python
print(df1.columns.tolist()) 
print(df2.columns.tolist())
```

```python
df2=df2.rename(columns={'TYPE':'Description','HUNDRED_BLOCK':'Location','NEIGHBOURHOOD':'Neighborhood'})
```

```python
print(df1.columns.tolist()) 
print(df2.columns.tolist())
```

```python
df_merged = pd.concat([df1, df2], ignore_index=True)
```

```python
df_merged
```

```python
value_counts = df_merged['Description'].value_counts()

print(value_counts)
```

```python
replace_dict = {
    'Theft from Vehicle': 'Theft',
    'Mischief': 'Mischief',
    'Break and Enter Residential/Other': 'Burglary',
    'LARCENY': 'Larency',
    'Other Theft': 'Theft',
    'COMMON ASSAULT': 'Assault',
    'BURGLARY': 'Burglary',
    'Theft of Vehicle': 'Theft',
    'LARCENY FROM AUTO': 'Larency',
    'Break and Enter Commercial': 'Burglary',
    'AGG. ASSAULT': 'Assault',
    'AUTO THEFT': 'Theft',
    'Theft of Bicycle': 'Theft',
    'Vehicle Collision or Pedestrian Struck (with Injury)': 'Accident',
    'ROBBERY - STREET': 'Robbery',
    'ROBBERY - COMMERCIAL': 'Robbery',
    'ASSAULT BY THREAT': 'Assault',
    'SHOOTING': 'Homicide',
    'ROBBERY - RESIDENCE': 'Robbery',
    'RAPE': 'Rape',
    'HOMICIDE': 'Homicide',
    'ROBBERY - CARJACKING': 'Robbery',
    'ARSON': 'Arson',
    'Vehicle Collision or Pedestrian Struck (with Fatality)': 'Accident'
}

df_merged['Description'] = df_merged['Description'].replace(replace_dict)
```
